```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree

# Importing the dataset/content/drive/MyDrive/DATA SCIENCE LAB/Social_Network_Ads.csv
dataset = pd.read_csv('/content/drive/MyDrive/iris.csv')
dataset
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
X=dataset.iloc[:,[2,3]]
y=dataset.iloc[:,4]
```

```python
 from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
X_test
```

```python
from sklearn.preprocessing import StandardScaler
sts=StandardScaler()
X_train=sts.fit_transform(X_train)
X_test=sts.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)
```

```python
X_test
```

```python
# Predicting the Test set results
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
# Plotting the Decision Tree graph
plt.figure(figsize=(10, 8))
plot_tree(classifier, feature_names=['petal.length', 'petal.width'], class_names=['Virginica', 'Setosa','Versicolor'], filled=True)
plt.show()
```
